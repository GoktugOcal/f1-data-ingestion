```python
import requests
import pandas as pd
import json
```

```python
season_year = 2024
r = requests.get(f"https://livetiming.formula1.com/static/{season_year}/Index.json")

season_data = json.loads(r.text[1:])
season_data
```

```python
for meeting in season_data["Meetings"]:
    for session in meeting["Sessions"]:
        try:
            url = "https://livetiming.formula1.com/static/" + session["Path"] + "Index.json"
            r = requests.get(url)
            print(url)
            print(r)
        except:
            pass
```
